## Предобработка данных
### Замена пропущенных значений
Перед работой с таблицей приведём названия колонок к общему виду.

```python
data.rename (columns = {'cityCenters_nearest': 'city_centers_nearest', 'parks_around3000': 'parks_around_3000', 'ponds_around3000': 'ponds_around_3000'}, inplace = True)
```

Посмотрим, в каких колонках и в каком объёме встречаются пропущенные значения.

```python
data.isna().sum()
```

Пропущенные значения в `ceiling_height` - отсутствие информации у пользователя при оформлении объявления. Значения можно заменить на медиану по всему датасету.

```python
data ['ceiling_height'].fillna (data ['ceiling_height'].median(), inplace = True)
```

Пропущенные значения в `floors_total` заполнять медианой будет не совсем корректно, т.к. может оказаться, что количество этажей в доме меньше, чем этаж квартиры. В данном случае лучше избавиться от этих значений, т.к. показатель общего количества этажей ещё будет нужен в иссследвании, а объём пропусков слишком мал, чтобы влиять на итоговые результаты (0,36% от всего датасета).

Пропущенные значения в `locality_name` также занимают слишком малую долю (0,21% от всего датасета), а месторасположение квартиры имеет значение при анализе - от этих данных лучше избавиться.

```python
data.dropna (subset = ['floors_total', 'locality_name'], inplace = True)
```

Для заполнения пропущенных значений в `living_area` и `kitchen_area` следует использовать средний показатель соотношения жилой площади и кухни к общей площади, предварительно их рассчитав.

```python
data ['living_area'].fillna (data ['total_area'] * (data ['living_area'] / data ['total_area']).mean(), inplace = True)
data ['kitchen_area'].fillna (data ['total_area'] * (data ['kitchen_area'] / data ['total_area']).mean(), inplace = True)
```

Пропущенные значения в `is_apartment` занимают практически всю колонку (88% от всего датасета). Разделить их на `True` или `False` невозможно без дополнительной информации. Данную колонку исключаем из датасета, т.к. она не несёт практической пользы в исследовании.

Также большой объём пропущенных значений в колонках `parks_nearest` (66% от всего датасета) и `ponds_nearest` (62% от всего датасета). Данные колонки также исключаем из датасета.

```python
data.drop (['is_apartment', 'parks_nearest', 'ponds_nearest'], axis = 1, inplace = True)
```

Пропущенные значения в `balcony` скорее всего говорят об отсутствии балкона в квартире, заменим их на 0.

Пропущенные значения в `parks_around3000` и `ponds_around3000` совпадают, вероятно, это ознаает отсутствие парков и водоёмов - заменим эти значенияна 0.

```python
fillna_columns = ['balcony', 'parks_around_3000', 'ponds_around_3000']
for i in fillna_columns:
    data [i].fillna (0, inplace = True)
```

Пропущенные значения в `days_exposition` могут говорить о ещё открытой сделке на момент формирования датасета, однако, если посмотреть на колонку `last_price` с пропущенными значениями в `days_exposition`, то видно, что цена на момент снятия публикации указана. Также эту теорию опровергает тот факт, что подобные строки встречаются с самого начала датасета (2014 год), т.е. скорее всего данные объявления действительно были закрыты и уже не актуальны. Это означает, что публикация была снята, но при этом по какой-то причине не посчиталось количество дней активного объявления. Вероятно, это ошибка системы, подсчитывающей данный показатель - пропуски оставим незаполненными.

```python
data [data ['days_exposition'].isna()][['first_day_exposition', 'last_price']].sort_values (by = 'first_day_exposition')
```

Пропущенные значения в колонках `airports_nearest` и `city_centers_nearest` тоже оставим без изменений. Значения в этих колонках практически полностью совпадают. Значения пропущены у большинства крупных населённых пунктов рядом с Санкт-Петербургом и занимают значительную часть датасета, однако, избавиться от этих колонок нельзя, т.к. значения понадобятся в дальнейшем исследовании.

Таким образом, мы избавились от основных пропущенных значений в датасете.

## Расчёты и добавление результатов в таблицу

В данном разделе мы проведём ряд расчётов для подготовки всех данных к последующему анализу:
* Расчитаем и добавим в таблицу: цену квадратного метра
* Расчитаем и добавим в таблицу: день недели, месяц и год публикации объявления
* Расчитаем и добавим в таблицу: этаж квартиры; варианты — первый, последний, другой
* Расчитаем и добавим в таблицу: соотношение жилой и общей площади, а также отношение площади кухни к общей

**Расчёт и добавление в таблицу цены квадратного метра**

```python
data ['square_price'] = (data ['last_price'] / data ['total_area']).astype ('int')
```

**Расчёт и добавление в таблицу дня недели, месяца и года публикации объявления**

```python
data ['weekday'] = data ['first_day_exposition'].dt.dayofweek
data ['month'] = data ['first_day_exposition'].dt.month_name()
data ['year'] = data ['first_day_exposition'].dt.year
```

**Расчёт и добавление в таблицу вариантов этажа квартиры**

```python
def floor_var (data):
    if data ['floor'] == 1:
        return 'первый'
    if data ['floor'] == data ['floors_total']:
        return 'последний'
    else:
        return 'другой'
data ['floor_var'] = data.apply (floor_var, axis = 1)
```

**Расчёт и добавление в таблицу соотношения жилой и общей площади, а также отношение площади кухни к общей**

```python
data ['living_ratio'] = round ((data ['living_area'] / data ['total_area'])*100, 1)
data ['kitchen_ratio'] = round ((data ['kitchen_area'] / data ['total_area'])*100, 1)

# также избавимся от колонок, которые не пригодятся в исследовании
data.drop (['first_day_exposition', 'total_images', 'studio', 'open_plan', 'airports_nearest', 'parks_around_3000', 'ponds_around_3000'], axis = 1, inplace = True)
data.head()
```

Напишем функцию `hist_and_describe`, которая будет возвращать гистограмму по заданным параметрам и выводить необходимые значения `describe`

```python
def hist_and_describe (_data, _bins, _range):
    _data.hist (bins = _bins, range = (_range), figsize = (10, 5))
    plt.show ()
    display (_data.describe ())
```

### Определение центра в Санкт-Петерубрге

Переведём значения `city_centers_nearest` в километры и округлим до целых чисел, полученные данные сохраним в колонку `center_km`. Затем посчитаем среднюю цену для каждого километра и сохраним в отдельную колонку `center_price`.

По полученным данным построим график, показывающий зависимость цены от удалённости от центра. Выборку цен за 1 м2 возьмём до 1 млн, чтобы сильно выбивающиеся значения не мешали анализу.
